# Зависит ли Стоимость игрока стоимость (Value) игрока от его "финтов" (Skill Moves)? 

 Как можно заметить, стоимость и правда зависит - медиана стоимости растет с ростом Skill Moves, растут и выбросы (например макс стоимость для игрока с Skill Moves = 1 - 72 миллиона евро, для игрока с Skill Moves = 2 - 63 миллиона, Skill Moves = 3 - 83.5 миллиона, Skill Moves = 4 уже 110.5 миллиона евро, ну и для игрока с Skill Moves = 5 цена уже 118.5 миллиона евро. 

```python
# переформатируем стоимости игроков:
def cut_the_price(price_in_str : str):
    #print("before:", price_in_str)
    if price_in_str[-1] == "K":
        return float(price_in_str[:-1]) 
    elif price_in_str[-1] == "M":
        return float(price_in_str[:-1]) * 1000
    else:
        return float(price_in_str) / 1000

df["Value"] = df['Value'].apply(lambda x : cut_the_price(x[1:]))
```

```python
px.box(df, x="Skill Moves", y="Value", title="Value's dependence on skill moves")
```

# Покажите распределеняи основных характеристик игроков: Рост, Вес, Возраст по позициям (Position), сделайте выводы...


```python
df = df.drop(index=df[df['Position'].isna()].index)
positions = df['Position'].unique() # 27
```

```python
def from_american_to_euro(height_am):
    foots = height_am.split("'")
    return int(int(foots[0]) * 30.48 + int(foots[1]) * 2.54)

df['Height'] = df['Height'].apply(lambda x : from_american_to_euro(x))
```

```python
def cut_lbs(bad_weight):
    return int(bad_weight[:-3])

df['Weight'] = df["Weight"].apply(lambda x : cut_lbs(x))
```

# Насколько средняя зарплата (Wage) игроков с рейтингом (Overall) 90+ больше средней зарплаты с рейтингом [80 - 90]? Визуализируйте возможную зависимость

```python
def make_wage_number(x):
    if (x[-1] == "K"):
        return int(x[1:-1]) * 1000
    else:
        return int(x[1:])
    
df["Wage"] = df["Wage"].apply(lambda wage : make_wage_number(wage))
mean_more_than_90_wage = df[df["Overall"] >= 90]["Wage"].mean()
mean_between_80_and_90_wage = df[(df["Overall"] >= 80)&(df["Overall"] < 90)]["Wage"].mean()
print("Разница между средней зарплатой у чуваков с рейтингом 90+ и чуваков с рейтингом 80-90 =", 
      mean_more_than_90_wage - mean_between_80_and_90_wage)
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(y=df[df['Overall'] >= 80].groupby("Overall")["Wage"].mean(), 
                     x=df[df['Overall'] >= 80].groupby("Overall")["Wage"].mean().index))
fig.update_layout(title="Mean wage over the 80+ overall rating",
                  xaxis_title='Overall Rating',
                  yaxis_title='Mean wage'
                 )
fig.show()
```
